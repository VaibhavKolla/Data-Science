```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y=  [20,42,53,48,42]
z=  [20,31,42,86,75]
```

```python
plt.plot(x,y,'g*')
```

```python
plt.plot(x,y , color = 'green',marker = 'D' ,linestyle = '',markersize = 10,alpha = 0.5) #alpha used for line/marker transparency
```

```python
plt.plot(x,y, label = 'Min')
plt.plot(y,z, label = 'Max')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Temprature")
plt.legend(loc = 'best', fontsize = 'small')
plt.grid()
```
