```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['Google','Amazon','Microsoft','Twitter']
revenue = [100,146,150,127]
profit = [40,20,35,52]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.2 ,revenue ,width = 0.4,label = 'revenue')
plt.xticks(xpos, company)
plt.xlabel('company(bln)')
plt.ylabel('revenue(bln)')
plt.title('US TECH STOCKS')
plt.bar(xpos+0.2, profit,width=0.4, label = 'Profit')
plt.legend()
```

```python
plt.barh(xpos-0.2 ,revenue ,label = 'revenue')
plt.yticks(xpos, company)
plt.xlabel('company(bln)')                  # For horizontal barchart width must be removed and xticks to yticks
plt.ylabel('revenue(bln)')
plt.title('US TECH STOCKS')
plt.barh(xpos+0.2, profit, label = 'Profit')
plt.legend()
```
