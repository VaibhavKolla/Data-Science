```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [100,85,90,83,72,112,115,120,111]
women = [112,73,65,86,86,92,88,64,96]
plt.hist([blood_sugar,women], color=['orange','green'],bins= 3 , rwidth = 0.987, label = ['Men', 'Women'],orientation = 'horizontal')
plt.legend(loc= "upper right" ,shadow= 2)
plt.xlabel('Sugar Range')
plt.ylabel('Total No. of Patients')
plt.title('Sugar Levels Of Individuals')
```
