```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values = [100,200,450,770,620]
exp_labels = ['Rent','Tax','Groceries','Bills','Invest/Savings']
```

```python
plt.pie(exp_values, labels = exp_labels,radius = 0.8 , autopct = '%0.2f%%', shadow = True , explode = [0,0,0,0.1,0], startangle= 10)
plt.axis('equal')
plt.show()
```
